### 3. データクリーニングと前処理

データの初期探索の結果、いくつかのデータ品質の問題が特定されました。これらの問題を解決し、データ分析に適した形に整形します。

**特定された課題点:**

*   **`InvoiceNo`のデータ型警告:** `InvoiceNo`列が混合型であるという警告が出ています。これは、数値と文字列が混在している可能性を示唆しており、キャンセルされた注文が 'C' で始まるなどのパターンが考えられます。この列はトランザクションを識別するために重要であり、一貫したデータ型に整理する必要があります。
*   **`InvoiceDate`のデータ型:** `InvoiceDate`が `object`型であり、日付として扱うためには `datetime`型に変換する必要があります。
*   **`CustomerID`の欠損値:** `CustomerID`列に約25%の欠損値があります。これは、ゲスト購入や登録されていない顧客による購入を示唆している可能性があります。顧客ベースの分析（例：RFM分析）を行う際にはこれらの行をどのように扱うか検討が必要です。
*   **`Quantity`と`UnitPrice`の不正な値:** 統計量から、`Quantity`と`UnitPrice`の最小値がそれぞれ1.0と0.001となっています。しかし、eコマースデータでは、数量が負の値（返品）や単価がゼロ（無料プロモーション品など）である場合があります。これらが分析にどのように影響するか考慮し、必要に応じて処理します。特に、負の数量は後で利益計算に影響するため、別途処理が必要です。

**データクリーニングのアクションプラン:**

1.  **キャンセル注文の特定と処理:** `InvoiceNo`が 'C' で始まる行をキャンセル注文として特定し、分析の目的に応じて除外するか、別途分析対象とするかを検討します。今回は、利益計算に不要な返品（マイナス数量）を考慮するため、キャンセル注文を適切に識別します。
2.  **データ型の変換:** `InvoiceDate`を `datetime`型に変換します。
3.  **欠損値の処理:** `CustomerID`の欠損値を持つ行は、顧客ごとの分析を行う上で障害となります。顧客IDがないトランザクションは、顧客の行動パターンを分析する際には除外しますが、全体売上分析には含めることも検討します。今回は、顧客ベースの分析のために欠損値のある行を除外します。
4.  **`Quantity`と`UnitPrice`の異常値処理:** `Quantity`が負の値の場合は返品とみなし、処理を検討します。また、`UnitPrice`が0の行も無料のアイテムなどとして扱いを検討します。今回は、正の数量と単価を持つ有効な取引のみを分析の対象とします。また、収益最大化を目的としているため、特に負の数量（返品）は利益を減少させる要因として注意深く扱います。
5.  **`TotalPrice`の計算:** 各トランザクションの総価格（`Quantity` × `UnitPrice`）を計算し、新しい列`TotalPrice`として追加します。

```python
print("### データクリーニング開始 ###\n")

# 1. InvoiceNoの処理とキャンセル注文の特定
# 'C'で始まるInvoiceNoを持つ行をキャンセル注文として特定
df['IsCancelled'] = df['InvoiceNo'].astype(str).str.startswith('C')
print(f"キャンセル注文（'C'で始まるInvoiceNo）の数: {df['IsCancelled'].sum()}")

# InvoiceNoから'C'を削除し、数値として扱えるようにする（必要であれば）
df['InvoiceNo'] = df['InvoiceNo'].astype(str).str.replace('C', '')

# 2. InvoiceDateをdatetime型に変換
# 以前の変換で多くのNaTが発生したため、より柔軟な方法で変換を試みる。
# dayfirst=True を使用して、日-月-年形式（英国などで一般的）を優先して解析する。
original_invoice_date_count = df['InvoiceDate'].count() # 変換前の非NULL数を取得
try:
    df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], dayfirst=True, errors='coerce')
    nat_count_after_conversion = df['InvoiceDate'].isnull().sum()
    if nat_count_after_conversion > 0:
        print(f"InvoiceDate変換後、依然として {nat_count_after_conversion}件のNaT（欠損日付）が残っています。")
        print("これはデータ内の日付フォーマットが非常に多様であるか、不正な日付データが含まれているためです。")
        print("この後の分析では、欠損日付の行は考慮しないか、別途処理が必要になります。")
    else:
        print("InvoiceDateがすべて正常にdatetime型に変換されました。")
except Exception as e:
    print(f"InvoiceDateのdatetime型変換中に予期せぬエラーが発生しました: {e}")
    # エラーが発生した場合は元のシリーズを保持するか、対応を検討
    # （ここではエラー発生時もNaTとして処理されるerrors='coerce'を使っているので、通常はここに到達しないはず）

print(f"InvoiceDateのデータ型: {df['InvoiceDate'].dtype}")
print(f"InvoiceDateの非NULL値の数: {df['InvoiceDate'].count()}件 (元の非NULL数: {original_invoice_date_count})")

# 3. CustomerIDの欠損値処理
# CustomerIDが欠損している行の数を表示
missing_customer_id_before = df['CustomerID'].isnull().sum()
print(f"CustomerID欠損値（処理前）: {missing_customer_id_before}件")

# CustomerIDが欠損している行を削除
# 顧客ベースの分析を行うため、顧客IDのないトランザクションは除外します。
original_rows = len(df)
df.dropna(subset=['CustomerID'], inplace=True)
print(f"CustomerID欠損値のある{original_rows - len(df)}件の行を削除しました。")

# CustomerIDを整数型に変換
df['CustomerID'] = df['CustomerID'].astype(int)

# 4. QuantityとUnitPriceの異常値処理
# Quantityが負の値（返品）の行の数を表示
negative_quantity_rows = df[df['Quantity'] < 0].shape[0]
print(f"Quantityが負の値の行数（返品）: {negative_quantity_rows}件")

# UnitPriceが0以下の行の数を表示
zero_or_negative_price_rows = df[df['UnitPrice'] <= 0].shape[0]
print(f"UnitPriceが0以下の行数: {zero_or_negative_price_rows}件")

# 収益に貢献しない（負の数量またはゼロ以下の単価）行を削除
# 正の数量と単価を持つ有効な取引のみを分析対象とする
initial_rows_for_analysis = len(df)
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
print(f"Quantity > 0 および UnitPrice > 0 ではない{initial_rows_for_analysis - len(df)}件の行を削除しました。")

# 5. TotalPriceの計算
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
print("TotalPrice列を追加しました。\n")

print("### クリーニング後のデータ情報 ###")
df.info()
print("\n### クリーニング後のデータの最初の5行 ###")
print(df.head())
```

### 5. データ分析

#### 5.1 全体的な売上・利益トレンドの把握

まずは、時間経過に伴う売上と利益の動向を視覚化し、マクロな視点でのビジネスパフォーマンスを理解します。これにより、季節性、成長、停滞などのパターンを特定し、将来の予測や戦略立案の基礎とします。

**分析の目的:**
*   月ごとの総売上と利益の推移を把握する。
*   売上と利益の季節性や傾向を特定する。
*   期間内のパフォーマンスのピークと谷を特定する。

**分析方法:**
1.  `InvoiceDate`から年と月を抽出し、月ごとの集計軸を作成します。
2.  月ごとの`TotalPrice`（売上）と推定利益（今回は`TotalPrice`を利益と仮定するか、別途利益率を考慮する）を計算します。
3.  これらのデータを折れ線グラフで視覚化します。

**注意点:** `InvoiceDate`が`NaT`の行は、この時間軸分析からは除外されます。

```python
import matplotlib.pyplot as plt
import seaborn as sns

# japanize_matplotlibがインストールされていない場合はインストール
try:
    import japanize_matplotlib
except ImportError:
    print("japanize_matplotlibをインストールします...")
    !pip install japanize-matplotlib --quiet # --quietで冗長な出力を抑制
    import japanize_matplotlib
    print("japanize_matplotlibがインストールされ、インポートされました。")

# InvoiceDateがNaTではない行のみを対象とする
df_time_series = df.dropna(subset=['InvoiceDate']).copy()

# 月ごとの売上を計算するために年と月を抽出
df_time_series['InvoiceMonth'] = df_time_series['InvoiceDate'].dt.to_period('M')

# 月ごとの総売上を計算
monthly_sales = df_time_series.groupby('InvoiceMonth')['TotalPrice'].sum().reset_index()
monthly_sales['InvoiceMonth'] = monthly_sales['InvoiceMonth'].dt.to_timestamp()

# グラフの作成
plt.figure(figsize=(15, 7))
sns.lineplot(x='InvoiceMonth', y='TotalPrice', data=monthly_sales, marker='o')
plt.title('月次総売上トレンド', fontsize=16)
plt.xlabel('年月', fontsize=12)
plt.ylabel('総売上 (GBP)', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("### 月次総売上トレンドの分析結果 ###")
print(monthly_sales.head())
print(monthly_sales.describe())
```

#### 5.2 顧客行動の分析（RFM分析）

月次売上トレンドから季節性を把握しましたが、さらに収益を最大化するためには、顧客の行動を深く理解することが不可欠です。RFM（Recency, Frequency, Monetary）分析は、顧客をその購買行動に基づいてセグメント化するための強力な手法です。

**RFM指標の定義:**
*   **Recency (最終購入日からの経過日数):** 顧客が最後に商品を購入してから何日経過したか。最近購入した顧客ほど、再購入の可能性が高いと考えられます。
*   **Frequency (購入頻度):** 顧客が一定期間内に何回購入したか。購入頻度が高い顧客ほど、ブランドへの忠誠度が高いと考えられます。
*   **Monetary (購入金額):** 顧客が一定期間内にいくら購入したか。購入金額が大きい顧客ほど、LTV（顧客生涯価値）が高い優良顧客であると考えられます。

**分析の目的:**
*   顧客をRFMスコアに基づいてセグメント化する。
*   各セグメントの特性を理解し、収益に貢献する優良顧客や、離反の可能性のある顧客を特定する。
*   特定の顧客セグメントに対するマーケティング戦略やアプローチを立案するための基礎情報とする。

**分析方法:**
1.  分析基準日を決定します（データ内の最新購入日よりも少し後の日付）。
2.  各顧客（`CustomerID`）について、Recency, Frequency, Monetaryを算出します。
    *   Recency: `CustomerID`ごとの最新の`InvoiceDate`から分析基準日までの日数を計算します。
    *   Frequency: `CustomerID`ごとのユニークな`InvoiceNo`の数を数えます。
    *   Monetary: `CustomerID`ごとの`TotalPrice`の合計を計算します。
3.  RFMスコアを計算するために、各指標をいくつかのランクに分割します（例: 1-5の5段階評価）。Recencyは値が小さいほど良い（スコアが高い）、FrequencyとMonetaryは値が大きいほど良い（スコアが高い）とします。
4.  RFMスコアを組み合わせて顧客セグメントを定義します（例: 優良顧客、要注目顧客、離反顧客など）。

**注意点:** `InvoiceDate`が`NaT`の行は、このRFM分析からは除外されます。また、`CustomerID`が欠損している行はデータクリーニングの段階で既に削除済みです。

```python
from datetime import datetime

print("### RFM分析開始 ###\n")

# InvoiceDateがNaTではない行のみを対象とする
df_rfm = df.dropna(subset=['InvoiceDate']).copy()

# 分析基準日を決定 (データ内の最新購入日+1日)
# データ内の最新購入日を特定
latest_purchase_date = df_rfm['InvoiceDate'].max()
analysis_date = latest_purchase_date + pd.Timedelta(days=1)
print(f"分析基準日: {analysis_date}\n")

# RFM指標の計算
rfm_data = df_rfm.groupby('CustomerID').agg(
    Recency=('InvoiceDate', lambda date: (analysis_date - date.max()).days), # 最新購入日からの日数
    Frequency=('InvoiceNo', 'nunique'),                                  # ユニークな注文数
    Monetary=('TotalPrice', 'sum')                                     # 総購入金額
).reset_index()

print("### RFMデータの最初の5行 ###")
print(rfm_data.head())
print("\n### RFMデータの基本統計量 ###")
print(rfm_data.describe())

# RFMスコアの計算 (各指標を5段階にランク付け)
# Recency: 日数が少ないほど高スコア (逆順)
rfm_data['R_Score'] = pd.qcut(rfm_data['Recency'], 5, labels=[5, 4, 3, 2, 1])
# Frequency: 回数が多いほど高スコア
# duplicates='drop' を適用し、生成されるカテゴリを数値に変換してスコアとする
rfm_data['F_Score'] = pd.qcut(rfm_data['Frequency'], 5, duplicates='drop')
rfm_data['F_Score'] = rfm_data['F_Score'].cat.codes + 1 # 0から始まるコードに1を加算して1ベースのスコアにする
# Monetary: 金額が多いほど高スコア
rfm_data['M_Score'] = pd.qcut(rfm_data['Monetary'], 5, labels=[1, 2, 3, 4, 5])

# RFMスコアを結合してセグメントを作成
rfm_data['RFM_Score'] = rfm_data['R_Score'].astype(str) + rfm_data['F_Score'].astype(str) + rfm_data['M_Score'].astype(str)

# RFMセグメントの定義 (例)
# (詳細なセグメンテーションは後ほど、今回は代表的なものに絞る)
def rfm_segment(row):
    if row['R_Score'] == 5 and row['F_Score'] == 5 and row['M_Score'] == 5:
        return '優良顧客 (Champions)'
    elif row['R_Score'] == 5 and row['F_Score'] == 5:
        return 'ロイヤル顧客 (Loyal Customers)'
    elif row['R_Score'] == 5 and row['M_Score'] == 5:
        return '高価値新規顧客 (New High-Value Customers)'
    elif row['R_Score'] == 5 and row['F_Score'] >= 3:
        return '有望顧客 (Promising)'
    elif row['R_Score'] >= 4 and row['F_Score'] >= 4:
        return '要活性化顧客 (Need Attention)'
    elif row['R_Score'] <= 2 and row['F_Score'] <= 2:
        return '離反危険顧客 (At Risk)'
    else:
        return 'その他 (Others)'

rfm_data['RFM_Segment'] = rfm_data.apply(rfm_segment, axis=1)

print("\n### RFM分析結果の最初の5行 ###")
print(rfm_data.head())

print("\n### RFMセグメントごとの顧客数 ###")
print(rfm_data['RFM_Segment'].value_counts())

# RFMセグメントごとの平均Monetary (売上貢献) を可視化
plt.figure(figsize=(12, 6))
sns.barplot(x=rfm_data['RFM_Segment'].value_counts().index, y=rfm_data['RFM_Segment'].value_counts().values)
plt.title('RFMセグメントごとの顧客数', fontsize=16)
plt.xlabel('RFMセグメント', fontsize=12)
plt.ylabel('顧客数', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x='RFM_Segment', y='Monetary', data=rfm_data.groupby('RFM_Segment')['Monetary'].mean().reset_index().sort_values(by='Monetary', ascending=False))
plt.title('RFMセグメントごとの平均購入金額', fontsize=16)
plt.xlabel('RFMセグメント', fontsize=12)
plt.ylabel('平均購入金額 (GBP)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### 6. 提案：RFM分析に基づくリテンション施策

RFM分析により「離反危険顧客 (At Risk)」セグメントが特定されました。これらの顧客は、Recency（最終購入日からの経過日数）とFrequency（購入頻度）が低い傾向にあり、このままでは顧客リストから離脱してしまう可能性が高いグループです。

「収益最大化」というKGIを達成するためには、新規顧客獲得だけでなく、既存顧客の維持、特に離反しそうな顧客の再活性化が非常に重要です。以下に、具体的なリテンション施策を提案します。

#### 離反危険顧客 (At Risk) へのリテンション施策

**特徴:** 最近の購入がなく、購入頻度も低い顧客。
**目的:** 顧客との接点を再構築し、再購入を促す。

1.  **パーソナライズされた再エンゲージメントメール/プッシュ通知**
    *   **内容:**
        *   「しばらくご利用がないようです」といった、顧客の状況に合わせたメッセージ。
        *   過去の購入履歴に基づいたおすすめ商品の提案（例：前回購入した商品のアップグレード版、関連商品）。
        *   限定的な割引クーポンや送料無料などの特典（例：「〇〇様限定！10%オフクーポン」）。
        *   人気商品の紹介や新着情報の提供。
    *   **現場の泥臭いアクション:** 顧客データから最終購入日を抽出し、一定期間（例：3ヶ月～6ヶ月）購入がない顧客を自動的にリストアップするシステムを構築。マーケティングオートメーションツールと連携し、ステップメールを自動配信する。
    *   **ROI意識:** クーポン利用率、メール開封率、クリック率、そして再購入率をKPIとして追跡し、施策の効果を測定する。クーポン費用対効果を算出する。

2.  **アンケートを通じたニーズ把握**
    *   **内容:** 「なぜ最近ご利用がないのか」「何か不満はなかったか」など、簡単なアンケートを実施。アンケート回答者には、謝礼として少額のポイントや次回利用可能な割引券を提供する。
    *   **現場の泥臭いアクション:** 離反顧客リストから無作為に顧客を抽出し、アンケートシステム（例：Google Forms, SurveyMonkey）でアンケートを実施。回答内容を分析し、サービス改善のヒントを得る。
    *   **ROI意識:** アンケート回答率、アンケートからのサービス改善点抽出、およびそれによる将来的な顧客離反率低下への寄与度（定性・定量の両面で評価）。

3.  **限定的なコミュニティへの招待**
    *   **内容:** 顧客ロイヤルティを高めるためのSNSグループや限定イベントへの招待。ブランドへの再関心を喚起する。
    *   **現場の泥臭いアクション:** 離反危険顧客の中からエンゲージメントが高まりそうな顧客を選定し、クローズドなSNSグループ（例：Facebookグループ、LINEオープンチャット）やオンラインイベントに招待状を送付する。
    *   **ROI意識:** コミュニティ参加率、コミュニティ内での顧客エンゲージメント、そしてコミュニティ参加後の再購入率の増加。

4.  **カスタマーサポートからのパーソナルなフォロー**
    *   **内容:** 特に高額商品を購入していたり、過去に複数回購入している顧客に対しては、電話やパーソナルなメッセージで状況を確認し、困りごとがないか、再購入を妨げている要因がないかを聞き出す。
    *   **現場の泥臭いアクション:** 離反危険顧客の中から、特にMonetaryスコアが高い顧客を抽出し、カスタマーサポートチームが個別に連絡を取るためのリストを作成。スクリプトを用意し、顧客の反応を記録する。
    *   **ROI意識:** 個別フォロー後の再購入率、顧客満足度の向上、顧客単価の維持。


これらの施策を組み合わせ、PDCAサイクルを回しながら効果測定を行い、継続的に改善していくことが収益最大化への道筋となります。

#### 5.3 商品パフォーマンス分析

顧客行動の分析（RFM分析）に続き、今度は商品の視点から売上と利益の最大化を目指します。どの商品が最も売上に貢献しているのか、または改善の余地があるのかを特定することは、在庫管理、プロモーション、商品開発戦略において非常に重要です。

**分析の目的:**
*   売上上位の商品（ベストセラー）と売上下位の商品（死に筋商品）を特定する。
*   商品の特性（単価、販売数量）と売上貢献度の関係を理解する。
*   特定の商品のプロモーション強化、在庫最適化、または廃止の検討の基礎とする。

**分析方法:**
1.  `Description`（商品名）ごとに`TotalPrice`の合計を計算し、売上貢献度を算出します。
2.  売上上位のN個の商品と、売上下位のN個の商品を抽出します。
3.  これらの商品を棒グラフなどで視覚化し、傾向を把握します。

**注意点:** `df`データフレームは、`Quantity`が負の値や`UnitPrice`が0以下の取引、`CustomerID`が欠損している行を除外したクリーンなデータを使用します。

```python
# 商品ごとの総売上を計算
product_sales = df.groupby('Description')['TotalPrice'].sum().reset_index()

# 売上上位10商品
top_10_products = product_sales.sort_values(by='TotalPrice', ascending=False).head(10)
print("### 売上上位10商品 ###")
print(top_10_products)

# 売上下位10商品 (ただし、TotalPriceが0より大きいものに限る)
# 完全に売上のない商品（TotalPrice=0）を除外することで、より意味のある「死に筋」を特定
bottom_10_products = product_sales[product_sales['TotalPrice'] > 0].sort_values(by='TotalPrice', ascending=True).head(10)
print("\n### 売上下位10商品 (売上実績のあるもの) ###")
print(bottom_10_products)

# --- 可視化 ---

# 売上上位10商品の棒グラフ
plt.figure(figsize=(14, 7))
sns.barplot(x='TotalPrice', y='Description', data=top_10_products, palette='viridis')
plt.title('売上上位10商品', fontsize=16)
plt.xlabel('総売上 (GBP)', fontsize=12)
plt.ylabel('商品名', fontsize=12)
plt.tight_layout()
plt.show()

# 売上下位10商品の棒グラフ
plt.figure(figsize=(14, 7))
sns.barplot(x='TotalPrice', y='Description', data=bottom_10_products, palette='magma')
plt.title('売上下位10商品 (売上実績のあるもの)', fontsize=16)
plt.xlabel('総売上 (GBP)', fontsize=12)
plt.ylabel('商品名', fontsize=12)
plt.tight_layout()
plt.show()
```

商品パフォーマンス分析の結果、売上上位の商品（ベストセラー）と売上下位の商品（死に筋商品）が明確になりました。これらの情報に基づき、収益を最大化するための具体的な施策を提案します。

#### ベストセラー商品への施策

**特徴:** 高い売上を継続的に生み出している商品。ビジネスの主要な収益源。
**目的:** 更なる売上拡大、顧客単価向上、または利益率の改善。

1.  **関連商品のクロスセル・アップセル推奨**
    *   **内容:** ベストセラー商品を購入した顧客に対し、その商品と関連性の高い他の商品を推薦する。または、より高機能・高価格な上位商品を提案する。
    *   **現場の泥臭いアクション:** Eコマースサイトの商品詳細ページやカートページに「この商品を買った人はこちらも見ています」「一緒に購入されることが多い商品」といったレコメンデーション機能を強化。メールマーケティングで、購入履歴に基づいたパーソナライズされた商品提案を行う。
    *   **ROI意識:** レコメンデーションからの購入率、平均注文額（AOV）の向上。レコメンデーションエンジンの改善効果を売上増で評価。

2.  **限定バンドル販売・セット販売の強化**
    *   **内容:** ベストセラー商品と相性の良い商品をセットにして、通常よりお得な価格で提供する。これにより、複数の商品を同時に購入してもらい、顧客単価を向上させる。
    *   **現場の泥臭いアクション:** 期間限定の特別セットを企画し、Webサイトの目立つ位置やSNS広告でプロモーションを行う。在庫過多の関連商品をバンドルに含めることで、在庫消化も図る。
    *   **ROI意識:** バンドル販売によるセット購入率、平均注文額、在庫回転率の改善。単体販売とセット販売の利益率比較。

3.  **プレミアム版・限定版の展開**
    *   **内容:** ベストセラー商品のデザインや素材、機能などをアップグレードしたプレミアム版や、特定のイベントに合わせた限定版を投入し、高単価での販売機会を創出する。
    *   **現場の泥臭いアクション:** 顧客からのフィードバックや市場トレンドを分析し、どのようなプレミアム要素が顧客に響くかを特定。製造部門と連携し、限定的な生産を行う。先行予約キャンペーンなどを実施する。
    *   **ROI意識:** プレミアム版・限定版の販売個数、単価、およびそれが全体の売上・利益に与える影響。

#### 死に筋商品への施策

**特徴:** 売上が極めて低い、またはほとんどない商品。在庫コストや管理コストの負担が大きい。
**目的:** 在庫の消化、損失の最小化、または廃止の検討。

1.  **大幅な価格改定・クリアランスセール**
    *   **内容:** 在庫過多の死に筋商品を対象に、原価割れ覚悟で大幅な割引を行い、早期に在庫を消化する。シーズンオフ商品やデザインの古い商品などが対象。
    *   **現場の泥臭いアクション:** 四半期ごとなどにクリアランスセールを定期的に開催し、特設ページやメルマガで大々的に告知。実店舗がある場合は、店舗でのアウトレット販売も検討する。
    *   **ROI意識:** 在庫滞留期間の短縮、倉庫費用の削減、現金の回収。最終的な売却損益の把握と最小化。

2.  **バンドル販売・おまけとしての活用**
    *   **内容:** ベストセラー商品や人気商品を購入した顧客に対し、死に筋商品をおまけとして無料で提供したり、少額でバンドルに含めたりする。これにより、顧客満足度を向上させつつ、死に筋商品を捌く。
    *   **現場の泥臭いアクション:** 「今だけ！〇〇ご購入で限定〇〇プレゼント！」といったキャンペーンを企画。倉庫でのピッキング効率も考慮し、関連性の高いベストセラー商品とのバンドルを優先する。
    *   **ROI意識:** おまけ戦略によるベストセラー商品の販売促進効果、顧客満足度の向上、死に筋商品の在庫削減効果。

3.  **商品情報の見直し・プロモーション強化**
    *   **内容:** 売上が低い原因が認知不足や情報不足にある場合、商品写真の改善、詳細な商品説明の追加、利用シーンの提案などを行い、商品の魅力を再アピールする。ターゲット層を見直した広告出稿も検討。
    *   **現場の泥臭いアクション:** Eコマースサイトの担当者が死に筋商品のページを徹底的に見直し、改善点を洗い出す。Webデザイナーやコンテンツライターと連携して、魅力を最大限に引き出すコンテンツを作成する。SNSでの紹介投稿など、費用対効果の高いプロモーションを試す。
    *   **ROI意識:** ページ閲覧数、転換率（CVR）の変化。プロモーション費用に対する売上増。

4.  **商品ラインナップからの廃止検討**
    *   **内容:** 上記の施策を講じても改善が見られない場合、その商品が市場ニーズに合っていない、またはビジネスモデルにそぐわないと判断し、生産・販売を終了する。
    *   **現場の泥臭いアクション:** 定期的な商品ポートフォリオの見直し会議で、売上データ、利益率、在庫状況、市場トレンドなどを総合的に評価し、廃止候補を選定する。製造元やサプライヤーとの契約関係を確認し、廃止にかかるコストを算出する。
    *   **ROI意識:** 不採算商品の廃止による在庫コスト、管理コスト、機会損失の削減効果。その後の新規商品開発へのリソース再配分。

## 🛠️ データ準備と初期探索

1.  `archive.zip` を解凍し、`EcommerceVisualization.csv` を特定。
2.  初期読み込みでエンコーディングエラー(`'utf-8' codec can't decode byte 0xa3`)が発生したため、`shift_jis` エンコーディングで再読み込みを実施。
3.  以下の課題が特定されました:
    -   `InvoiceNo` 列の混合型警告 (`DtypeWarning`)
    -   `InvoiceDate` が `object` 型
    -   `CustomerID` に約25%の欠損値
    -   `Quantity` と `UnitPrice` に潜在的な不正値（負の値や0以下の値）

## 🧹 データクリーニングと前処理

収益最大化のための分析基盤を確立するため、以下のデータクリーニングと前処理を行いました。

-   **`InvoiceNo` の処理:** キャンセルを示す 'C' を除去し、`IsCancelled` フラグを追加。
-   **`InvoiceDate` の型変換:** `datetime`型への変換を試みましたが、データフォーマットの多様性により多数の`NaT`（欠損日付）が発生 (`dayfirst=True`を適用しても約57%が`NaT`）。時間ベースの分析では、`NaT`ではない行を対象としました。
-   **`CustomerID` の欠損値処理:** 顧客ベースの分析のため、`CustomerID`が欠損している行（131,326行）を削除。
-   **`Quantity` と `UnitPrice` の異常値処理:** `Quantity`が負の値、または`UnitPrice`が0以下の行は分析から除外（今回のデータでは該当する行は0件）。正の数量と単価を持つ有効な取引のみを対象。
-   **`TotalPrice` 列の追加:** `Quantity` × `UnitPrice` で各トランザクションの総売上を示す `TotalPrice` 列を新規追加。

2.  **限定バンドル販売・セット販売の強化:**
    -   **泥臭いアクション:** ベストセラーと関連商品をセットで提供。在庫過多の関連商品をバンドルに含め在庫消化も図る。
    -   **ROI意識:** バンドル販売によるセット購入率、AOV、在庫回転率の改善。
3.  **プレミアム版・限定版の展開:**
    -   **泥臭いアクション:** 顧客フィードバックやトレンドを分析し、高単価のプレミアム版や限定版を企画・製造。先行予約キャンペーンを実施。
    -   **ROI意識:** プレミアム版・限定版の販売個数、単価、全体売上・利益への貢献。

#### 死に筋商品への施策

1.  **大幅な価格改定・クリアランスセール:**
    -   **泥臭いアクション:** 定期的にクリアランスセールを開催。特設ページやメルマガで告知し、早期に在庫を消化。
    -   **ROI意識:** 在庫滞留期間短縮、倉庫費削減、現金回収。売却損益の最小化。
2.  **バンドル販売・おまけとしての活用:**
    -   **泥臭いアクション:** 人気商品購入時におまけとして死に筋商品を提供、またはバンドルに含める。「今だけ！〇〇ご購入で限定〇〇プレゼント！」キャンペーン。
    -   **ROI意識:** ベストセラー販売促進効果、顧客満足度向上、死に筋在庫削減効果。
3.  **商品情報の見直し・プロモーション強化:**
    -   **泥臭いアクション:** ECサイトの商品ページを改善（写真、説明文、利用シーン）。ターゲット層を見直した広告出稿やSNSでの紹介。
    -   **ROI意識:** ページ閲覧数、転換率（CVR）の変化。プロモーション費用対効果。
4.  **商品ラインナップからの廃止検討:**
    -   **泥臭いアクション:** 定期的な商品ポートフォリオ見直し会議で、売上、利益率、在庫、トレンドを総合評価し、廃止候補を選定。廃止にかかるコストを算出。
    -   **ROI意識:** 不採算商品廃止によるコスト削減、新規商品開発へのリソース再配分。
